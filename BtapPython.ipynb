```python
#Ex15
def find_sub(x, sub):
    pos = []
    for i in range(len(x)):
        k = i + len(sub)
        if x[i: k] == sub:
            pos.append(i)
    return pos

x = input('Input:')
sub = input("You want to find:")
print(find_sub(x, sub))
```
